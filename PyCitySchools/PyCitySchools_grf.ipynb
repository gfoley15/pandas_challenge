## Summary.
The school district analysis provides an overview of key data points for each school in the district, including student count, school type, and budget. 
It averages math and reading scores and calculates percent passing rates to evaluate the performance of each school.
Further analysis is conducted to examine scores and passing rates based on spending/budget tier, school size, and school type. 
This in-depth examination allows for conclusions to be drawn regarding the impact of school budget, size, and type on academic success, as measured by scores and passing rate
## Conclusions.

1. Charter schools consistently outperform District schools in math/reading scores and passing rates (overall passing rates included).
The overall passing rate is significantly higher at Charter schools than District schools - 90.43% to 53.67% respectiely. 
Also, when broken down to the individual school level, the Top 5 schools in the district are Charter schools and the bottom 5 are Di trict.
As a whole, Charter school students are having more academic success when compared to the District school s udents.
This indicates Charter schools may have more effective academic curriculum or more robust resourcing for cl ssrooms.
A possible reason for Charter schools having a positive correlation with higher average scores and passing rates can be found when reviewing spendi g budget.
Despite lower total spending budgets, Charter schools allocate their resources with a heavier focus on academic success, resulting in higher student a hievement.
While District schools had higher spending budgets, the allocation of resources may not be as targeted towards academic success as in Chartr schools. 
In sum, school type has a significant impact on academic performance and how budgets are utilized is more critical than the total amount of spending budget in influencing academic outcomes.


2. Small to medium sized schools have higher average scores and passing rates than large schools. Small schools specifically have the highest average scores for m th and reading.
This suggests small and medium sized schools may provide an environment more conducive to learning and acad mic performance.
Small to medium sized schools have less students to manage, and therefore may be able to cater curriculums or programs to student's learning styles. This helpful attribute could postiively impact
student's ability to focus, learn and perform at a high level. Since larger schools have more students and potentially limited staff resources per student compared to small or me ium sized schools,
individualized curriculums or smaller group learning is not a viable option. Without small/medium sized group learning, education quality and student scores are negateively impacted. Large schools c uld consider moving
to a new curriculum that matches how small and medium sized schools approach their curriculum as well as reviewing the potential need to hire more staff members to provide more individualized supp rt for all students.
There is also a relationship between school size and school type. All small and medium sized schools are Charter schools, thus the points made above also apply. In sum, a smaller school size positively
impacts academic performance suggesting that schools should consider the benefits of smaller class sizes and tailored educational approachupport students.

```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Create DataFrame for single dataset
school_df = pd.DataFrame(school_data_complete)
school_df.sample(5)
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_df['school_name'].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_df['Student ID'].count()
student_count
```

```python
# Calculate the total budget - done
budget_dedupe = school_df.drop_duplicates(subset=['school_name', 'budget']) # dedupe DataFrame based on school_name, budget
budget_df = budget_dedupe[['school_name', 'budget']] # create new DataFrame with only deduped school_name, budget values
total_budget = budget_df['budget'].sum() # sum up the budget values for each school to get total budget
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_df['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_df['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_df[(school_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = school_df[(school_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading 
passing_math_reading_count = school_df[(school_df["math_score"] >= 70) & (school_df["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame 
# Creating dictionary as DataFrame
district_summary = pd.DataFrame({
    'Total Schools': [school_count],
    'Total Students': [student_count],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_rate]
})

# Formatting columns
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

#Display DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the type per school from school_data
school_types = school_data.set_index(["school_name"])["type"]
school_types = pd.DataFrame(school_types)
school_types
```

```python
# Calculate the total student count per school from school_data
per_school_counts_df = school_df[['school_name', 'Student ID']]
per_school_counts = per_school_counts_df.groupby('school_name').count()['Student ID']
per_school_counts = pd.DataFrame(per_school_counts)
per_school_counts = per_school_counts.rename(columns = {'Student ID': 'Student Count'})
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school from school_data
per_school_budget_dedupe = school_df.drop_duplicates(subset=['school_name', 'budget']) # dedupe DataFrame based on school_name, budget
per_school_budget_df = per_school_budget_dedupe[['school_name', 'budget']] # create new DataFrame with only deduped school_name, budget values
per_school_budget = per_school_budget_df.groupby('school_name').sum() # sum up the budget values for each school to get total budget per school

per_school_capita = per_school_budget['budget'] / per_school_counts['Student Count'] # divide budget by student count(previously defined above) to get capita spending per school
per_school_budget["per capita"] = per_school_capita # adding per capita to per_school_budget dataframe
per_school_budget
```

```python
# Calculate the average test scores per school from school_data_complete
average_test_scores = school_df[['school_name', 'math_score', 'reading_score']].groupby('school_name').mean()
average_test_scores = average_test_scores.rename(columns = {'math_score':'avg_math_score', 'reading_score':'avg_reading_score'})
average_test_scores
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete
students_passing_math = school_df.loc[(school_df["math_score"] >= 70)]
school_students_passing_math = students_passing_math.groupby('school_name').count()["Student ID"]
school_students_passing_math = pd.DataFrame(school_students_passing_math)
school_students_passing_math = school_students_passing_math.rename(columns = {'Student ID': 'students_passed_math'})
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete 
students_passing_reading = school_df.loc[(school_df["reading_score"] >= 70)]
school_students_passing_reading = students_passing_reading.groupby('school_name').count()["Student ID"]
school_students_passing_reading = pd.DataFrame(school_students_passing_reading)
school_students_passing_reading = school_students_passing_reading.rename(columns = {'Student ID': 'students_passed_reading'})
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_df.loc[(school_df["reading_score"] >= 70) & (school_df["math_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).count()['Student ID']
school_students_passing_math_and_reading = pd.DataFrame(school_students_passing_math_and_reading)
school_students_passing_math_and_reading = school_students_passing_math_and_reading.rename(columns = {'Student ID':'students_passed_math_&_reading'})
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
# Calculate passing rates for math, reading, and overall passing
per_school_passing_math = (students_passing_math.groupby('school_name').count()["Student ID"] / per_school_counts_df.groupby('school_name').count()['Student ID']) * 100
per_school_passing_reading = (students_passing_reading.groupby('school_name').count()["Student ID"] / per_school_counts_df.groupby('school_name').count()['Student ID']) * 100
overall_passing_rate = (students_passing_math_and_reading.groupby(["school_name"]).count()['Student ID'] / per_school_counts_df.groupby('school_name').count()['Student ID']) * 100

# Create DataFrames for each passing rate
passing_math_df = pd.DataFrame({"% Passing Math": per_school_passing_math})
passing_reading_df = pd.DataFrame({"% Passing Reading": per_school_passing_reading})
passing_overall_df = pd.DataFrame({"% Passing Overall": overall_passing_rate})

# Merge DataFrames into single DataFrame
passing_rates_df = pd.concat([passing_math_df, passing_reading_df, passing_overall_df], axis = 1)

# Disply single DataFrame
passing_rates_df
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary1 = pd.merge(
    school_types,
    per_school_counts,
    how="left", 
    on=["school_name"])

per_school_summary2 = pd.merge(
    per_school_budget,
    average_test_scores,
    how="left", 
    on=["school_name"])
    
per_school_summary3 = pd.merge(
    school_students_passing_math,
    school_students_passing_reading,
    how="left", 
    on=["school_name"])

per_school_summary4 = pd.merge(    
    school_students_passing_math_and_reading,
    passing_rates_df,
    how="left", 
    on=["school_name"])

per_school_summary5 = pd.merge(    
    per_school_summary1,
    per_school_summary2,
    how="left", 
    on=["school_name"])

per_school_summary6 = pd.merge(    
    per_school_summary3,
    per_school_summary4,
    how="left", 
    on=["school_name"])

per_school_summary = pd.merge(    
    per_school_summary5,
    per_school_summary6,
    how="left", 
    on=["school_name"])


# Rename columns
per_school_summary = per_school_summary.rename(columns = {
    'type':'School Type',
    'budget':'Total School Budget',
    'per capita':'Per Student Budget',
    'avg_math_score':'Average Math Score',
    'avg_reading_score':'Average Reading Score',
    'students_passed_math':'Passed Math Stud. Count',
    'students_passed_reading':'Passed Reading Stud. Count',
    'students_passed_math_&_reading':'Passed Math & Reading Stud. Count', 
})

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:.2f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:.2f}".format)
per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:.2f}%".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:.2f}%".format)
per_school_summary["% Passing Overall"] = per_school_summary["% Passing Overall"].map("{:.2f}%".format)
per_school_summary = per_school_summary.sort_values("school_name")

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary.copy()
school_spending_df = pd.DataFrame(school_spending_df)
```

```python
# Converting, re-establishing variables as strings to utilize str.replace below
school_spending_df['Per Student Budget'] = school_spending_df['Per Student Budget'].astype(str)
school_spending_df['% Passing Math'] = school_spending_df['% Passing Math'].astype(str)
school_spending_df['% Passing Reading'] = school_spending_df['% Passing Reading'].astype(str)
school_spending_df['% Passing Overall'] = school_spending_df['% Passing Overall'].astype(str)
school_spending_df['Average Math Score'] = school_spending_df['Average Math Score'].astype('float')
school_spending_df['Average Reading Score'] = school_spending_df['Average Reading Score'].astype('float')
```

```python
# Converting string variables to float, removing $ and % signs from previous formatting
school_spending_df['Per Student Budget'] = school_spending_df['Per Student Budget'].str.replace("$","").astype('float')
school_spending_df['% Passing Math'] = school_spending_df['% Passing Math'].str.replace("%","").astype('float')
school_spending_df['% Passing Reading'] = school_spending_df['% Passing Reading'].str.replace("%","").astype('float')
school_spending_df['% Passing Overall'] = school_spending_df['% Passing Overall'].str.replace("%","").astype('float')
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per Student Budget'],
                                                            spending_bins, 
                                                             labels=labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Overall"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    'Average Math Score': spending_math_scores, 
    'Average Reading Score': spending_reading_scores, 
    '% Passing Math': spending_passing_math, 
    '% Passing Reading': spending_passing_reading,
    '% Passing Overall': overall_passing_spending
})

# Display results
spending_summary
```
